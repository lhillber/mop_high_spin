```python
import numpy as np
import matplotlib.pyplot as plt
import magnetics
from importlib import reload
reload(magnetics)
from magnetics import Field
```

```python
geometry = dict(
    config="Nline",
    Nline=6,
    taper=0.573,
    a = 0.5-2.5*np.tan(0.573*np.pi/180),
    L = 5,
    d = 0.086,
    I=100,
    ang=0,
    n=[0,0,1],
    r0=[0.0,0,0],
    meshspec=[[-0.1,0.1,200],[-0.1,0.1,200],[-3,3,200]]
)
field = Field(geometry, save=False, recalc=True)

for z in [-1.23, 0, 1.21]:
    r = np.linspace(-0.05, 0.05, 500)
    zs = np.ones_like(r)*z
    rs =np.zeros((500, 3))
    rs[:,0]=r
    rs[:,2]=zs
    plt.plot(r, 1e12*field.normB_interp(rs))
plt.show()



r = np.linspace(-3, 3, 500)
rs =np.zeros((500, 3))
x = 0.05 * np.ones_like(r)
rs[:,2]=r
rs[:,0]=x
plt.plot(r, 1e12*field.normB_interp(rs))
```
